```
import requests

def download_file_from_google_drive(id, destination):
  URL = "https://docs.google.com/uc?export=download"

  session = requests.Session()

  response = session.get(URL, params = { 'id' : id }, stream = True)
  token = get_confirm_token(response)

  if token:
      params = { 'id' : id, 'confirm' : token }
      response = session.get(URL, params = params, stream = True)

  save_response_content(response, destination)    

def get_confirm_token(response):
  for key, value in response.cookies.items():
      if key.startswith('download_warning'):
          return value

  return None

def save_response_content(response, destination):
  CHUNK_SIZE = 32768

  with open(destination, "wb") as f:
      for chunk in response.iter_content(CHUNK_SIZE):
          if chunk: # filter out keep-alive new chunks
              f.write(chunk)

if __name__ == "__main__":
  file_id = '1lgA9HZ9offdfCPpjjUNtK3-vI8jigcv8'
  destination = 'classification/data/cnews/cnews.train.txt'
  download_file_from_google_drive(file_id, destination)
```

```
def get_train_test_set():
  train_set=[]
  test_set=[]
  with open("classification/data/rec_news/user_click_data.txt") as f:
    for line in f.readlines():
      data = line.split("\t")
      try:
        time_news = int(data[2])
        if time_news>1395273600.0:#前20天当做训练集
          test_set.append(data)
        else:
          train_set.append(data)
      except Exception:
        pass
  return train_set,test_set
```

```
train_set,test_set=get_train_test_set()
```

```
#给user,news一个自然数ID

def get_user_news_ID():
  user_id={}
  news_id={}
  with open("classification/data/rec_news/user_click_data.txt") as f:
    userid=0
    newsid=0
    for line in f.readlines():
      data = line.split("\t")
      if data[0] not in user_id:
        user_id[data[0]]=userid
        userid+=1
      if data[1] not in news_id:
        news_id[data[1]]=newsid
        newsid+=1
      
  return user_id,news_id
```

```
user_id,news_id=get_user_news_ID()
```

```
# user_news矩阵
import numpy as np

def get_user_news_matrix():
  user_news_m=np.zeros([len(user_id),len(news_id)])
  for data in train_set:
    uid = user_id[data[0]]
    nid = news_id[data[1]]
    user_news_m[uid,nid]+=1
  return user_news_m
```

```
user_news_m=get_user_news_matrix()
```

# NMF

```
from sklearn.decomposition import NMF

nmf = NMF(n_components=10)  
user_distribution = nmf.fit_transform(user_news_m)  
item_distribution = nmf.components_
 
reconstruct_matrix = np.dot(user_distribution, item_distribution)  
filter_matrixs = user_news_m < 1e-6  # 小于0  

print('用户的主题分布：') 
print(user_distribution)  
print('物品的主题分布：')  
print(item_distribution)
print('重建矩阵，并过滤掉已经评分的物品：')
print(reconstruct_matrix*filter_matrixs)
```

# SVD

```
from scipy.sparse.linalg import svds  
from scipy import sparse


def vector_to_diagonal(vector):  
    """
    将向量放在对角矩阵的对角线上
    :param vector:
    :return:
    """
    if (isinstance(vector, np.ndarray) and vector.ndim == 1) or \
            isinstance(vector, list):
        length = len(vector)
        diag_matrix = np.zeros((length, length))
        np.fill_diagonal(diag_matrix, vector)
        return diag_matrix
    return None

RATE_MATRIX = user_news_m.astype('float')  
U, S, VT = svds(sparse.csr_matrix(RATE_MATRIX),  k=10, maxiter=200) # 2个隐主题  
S = vector_to_diagonal(S)
 
print('用户的主题分布：') 
print(U[0])  
print('奇异值：')
print(S) 
print('物品的主题分布：') 
print(VT) 
print('重建评分矩阵，并过滤掉已经评分的物品：')  
print(np.dot(np.dot(U, S), VT) * (RATE_MATRIX < 1e-6))
print(user_topic_m[0])
```

**至此有三种方式：**

1.  先预测文章主题，再根据用户的主题喜好和测试新闻的主题进行计算相似度
2. 通过NMF对user-news分解，然后根据用户分布和物品分布进行计算相似度
3.  通过SVD对user-news分解，根据用户的主题分布和物品的分布进行计算相似度
